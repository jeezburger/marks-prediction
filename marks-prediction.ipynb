```python
import pandas as pd

data = {
    'Hours': [11.95, 9.55, 7.03, 6.25, 5.28],
    'Scores': [36, 37, 31, 28, 27]
}
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Hours'], df['Scores'])
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['Hours']]  # features
y = df['Scores']   # labels

model = LinearRegression()
model.fit(X, y)
```

```python
import pandas as pd

model.predict(pd.DataFrame([[15]], columns=["Hours"]))
```

```python
def clamp_prediction(pred):
    return max(0, min(pred, 50))
```

```python
raw_pred = model.predict(pd.DataFrame([[11]], columns=["Hours"]))[0]  
clamped_pred = clamp_prediction(raw_pred)

print(f"Raw prediction: {raw_pred}")
print(f"Clamped prediction: {clamped_pred}")
```
